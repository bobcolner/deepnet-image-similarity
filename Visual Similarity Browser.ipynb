```python
import graphlab
def load_product_sim(catalog_name):
    # load images
    image_sframe = graphlab.load_sframe(catalog_name + '_images')
    # load image similarity matrix
    similarity = graphlab.load_sframe(catalog_name + '_sim.csv')
    return(image_sframe, similarity)

def show_similar(similarity, image_sframe, query_image=0):
    # show ranked similar images 
    graphlab.canvas.set_target('ipynb')
    ranked_images = similarity[similarity['query_label'] == query_image].sort('rank')
    for similar_img in ranked_images['reference_label']:
        graphlab.SArray([image_sframe['image'][similar_img]]).show()
```

```python
# catalog_name = 'mens_product'
catalog_name ='amazon_image'
image_sframe, similarity = load_product_sim(catalog_name)
```
